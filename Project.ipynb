```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
raw_data_count - cleaned_data_count
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample Size vs. Median')
```

It appears that the sample size has little to do with the median salary achieved after graduating from a specific major.

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample Size vs. Unemployment Rate')
```

```python
ax = recent_grads.plot('Full_time', 'Median', kind='scatter')
ax.set_title('Full Time vs. Median')
```

There is not a very strong relationship between the number of full-time employees and the median salary.

```python
ax = recent_grads.plot('ShareWomen', 'Unemployment_rate', kind='scatter')
ax.set_title('Share of Women vs. Unemployment Rate')
```

```python
ax = recent_grads.plot('Men', 'Median', kind='scatter')
ax.set_title('Men vs. Median')
```

```python
ax = recent_grads.plot('Women', 'Median', kind='scatter')
ax.set_title('Women vs. Median')
```

It appears that there may be a weak negative relationship between the number of women in a major and the median salary.
